### Export all haadf (or selected haadfs) to png

```python
haadf_stack = tf.imread('../../../primary_data/main_mosaic_5by6_haadf.tiff')
print(haadf_stack.shape)
```

```python
selection = [1,8]
for tile in selection:
    img = haadf_stack[tile,:,:]
    f,ax = plt.subplots(figsize=(2,2))
    plt.subplots_adjust(top = 1, bottom = 0, right = 1, left = 0, hspace = 0, wspace = 0)
    ax.imshow(img,cmap='gray')
    plt.margins(0,0)
    plt.axis('off')
    plt.savefig('../../secondary_data/Figure_5_HAADF%02d.png' % tile,dpi=300)
```
