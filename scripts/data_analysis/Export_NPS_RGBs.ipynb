```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
import sys

sys.path.append("../")
from functions_EDX import *
```

```python
spectrum_folder = '../../../primary_data/main_mosaic_6by5/Spectrum2D_PC10'
file_names = []
tmp = os.listdir(spectrum_folder)
for filename in tmp:
    if filename.endswith('.npz'):
        file_names.append(filename)

file_names = sorted(file_names)
files = [os.path.join(spectrum_folder,file_name) for file_name in file_names]
```

```python
for tile_idx,file in enumerate(files):
    spectrum = np.load(files[tile_idx])['spectrum_2D'].reshape((1024,1024,250))

    fig,ax = plt.subplots(figsize=(5,5))
    rgb = cv2.merge([normalize8(spectrum[:,:,4]), normalize8(spectrum[:,:,25]), normalize8(spectrum[:,:,28])])

    ax.imshow(rgb)
    plt.show()
```
